```python
import numpy as np
import json
import matplotlib.pyplot as plt
```

```python
pth = 'Data_to_Plot'

with open(pth + '/MOO.json', 'r') as f:
    RelVarConf = json.load(f)
RelVarConf = np.array(RelVarConf)

with open(pth + '/MOOSE.json', 'r') as f:
    RelVarEst = json.load(f)
RelVarEst = np.array(RelVarEst)

with open(pth + '/SH.json', 'r') as f:
    UnifConf = json.load(f)
UnifConf = np.array(UnifConf)

with open(pth + '/SHSE.json', 'r') as f:
    UnifEst = json.load(f)
UnifEst = np.array(UnifEst)

with open(pth + '/SHVar.json', 'r') as f:
    VarConf = json.load(f)
VarConf = np.array(VarConf)

with open(pth + '/SHVarSE.json', 'r') as f:
    VarEst = json.load(f)
VarEst = np.array(VarEst)

with open(pth + '/setup.json', 'r') as f:
    data = json.load(f)
```

```python
horizons = data['T']
_, iterations = np.shape(RelVarConf)

RelVarConf_mean = np.mean(RelVarConf, axis=1)
RelVarConf_std = np.std(RelVarConf, axis=1)
UnifConf_mean = np.mean(UnifConf, axis=1)
UnifConf_std = np.std(UnifConf, axis=1)
VarConf_mean = np.mean(VarConf, axis=1)
VarConf_std = np.std(VarConf, axis=1)
RelVarEst_mean = np.mean(RelVarEst, axis=1)
RelVarEst_std = np.std(RelVarEst, axis=1)
UnifEst_mean = np.mean(UnifEst, axis=1)
UnifEst_std = np.std(UnifEst, axis=1)
VarEst_mean = np.mean(VarEst, axis=1)
VarEst_std = np.std(VarEst, axis=1)
```

```python
RelVarConf_err = RelVarConf_std / np.sqrt(iterations) * 1.99
UnifConf_err = UnifConf_std / np.sqrt(iterations) * 1.99
VarConf_err = VarConf_std / np.sqrt(iterations) * 1.99
RelVarEst_err = RelVarEst_std / np.sqrt(iterations) * 1.99
UnifEst_err = UnifEst_std / np.sqrt(iterations) * 1.99
VarEst_err = VarEst_std / np.sqrt(iterations) * 1.99
```

```python
# Plot with error bars
plt.plot(horizons, RelVarEst_mean, marker = 'x', label='SHRVar', color='blue', markersize=6)
plt.plot(horizons, UnifEst_mean, marker = 'x', label='SH-z', color='green', markersize=6)
plt.plot(horizons, VarEst_mean, marker = 'x', label='SHVar-z', color='orange', markersize=6)
plt.plot(horizons, RelVarConf_mean, marker = 'o', label='SHRVar-c', color='blue', markersize=6)
plt.plot(horizons, UnifConf_mean, marker = 'o', label='SH-c', color='green', markersize=6)
plt.plot(horizons, VarConf_mean, marker = 'o', label='SHVar-c', color='orange', markersize=6)


# plt.errorbar(horizons, RelVarConf_mean, yerr=RelVarConf_err, ecolor='red', capsize=8)
# plt.errorbar(horizons, RelVarEst_mean, yerr=RelVarEst_err, ecolor='red', capsize=8)
# plt.errorbar(horizons, UnifConf_mean, yerr=UnifConf_err, ecolor='red', capsize=8)
# plt.errorbar(horizons, UnifEst_mean, yerr=UnifEst_err, ecolor='red', capsize=8)
# plt.errorbar(horizons, VarConf_mean, yerr=VarConf_err, color='orange', ecolor='red', capsize=8)
# plt.errorbar(horizons, VarEst_mean, yerr=VarEst_err, color='orange', ecolor='red', capsize=8)

plt.ylim([0.2,1])
# Customize the plot
plt.xlabel("Horizon")
plt.ylabel("Accuracy")
plt.legend(loc='lower right')
plt.grid(True)
# plt.savefig('Figs/ErrPlot.pdf')
plt.show()
```
